3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

```python
import os

folder1 = "part5"
folder2 = "part4"

def get_lexical_diversity(folder_path):
    all_tokens = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
                text = f.read()
                tokens = word_tokenize(text.lower())
                all_tokens.extend(tokens)
    unique_tokens = set(all_tokens)
    diversity = len(unique_tokens) / len(all_tokens)
    return diversity, len(all_tokens), len(unique_tokens)

div1, total1, unique1 = get_lexical_diversity(folder1)
div2, total2, unique2 = get_lexical_diversity(folder2)

print(f"Folder 1 ({folder1}): {total1} tokens, {unique1} unique, diversity = {div1:.4f}")
print(f"Folder 2 ({folder2}): {total2} tokens, {unique2} unique, diversity = {div2:.4f}")
```

5. Now, write a code block that depends on variables you set for your file directories in the previous code block(s). This codeblock should just select a set of tokens from your text based on their part of speech, either adjectives or verbs.  *(5 pts)* 

```python
def get_verbs(folder_path):
    all_verbs = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
                text = f.read()
                tokens = word_tokenize(text.lower())
                tagged = nltk.pos_tag(tokens)
                verbs = [word for word, tag in tagged if tag.startswith("VB")]
                all_verbs.extend(verbs)
    return all_verbs

verbs1 = get_verbs(folder1)
verbs2 = get_verbs(folder2)

print(f"Folder 1 verb count: {len(verbs1)}")
print(f"Folder 2 verb count: {len(verbs2)}")
```

6. Develop the code in the previous block by copying it into the next code block, and here:
   * Develop code to find out about the top 15 most ambiguous words in your adjective or verb token list, and
   * Write your output to a neatly formatted text file for reading. The text output file should provide a clear heading to provide a title for each corpus you're comparing. *(5 pts)*

```python
def get_ambiguous_words(verbs, top_n=15):
    freq = FreqDist(verbs)
    ambiguity = {}
    for word in freq.keys():
        synsets = wordnet.synsets(word, pos=wordnet.VERB)
        if synsets:
            ambiguity[word] = len(synsets)
    sorted_words = sorted(ambiguity.items(), key=lambda x: x[1], reverse=True)
    return sorted_words[:top_n]

ambiguous1 = get_ambiguous_words(verbs1)
ambiguous2 = get_ambiguous_words(verbs2)

# write to output file
with open("ambiguity_output.txt", "w", encoding="utf-8") as f:
    f.write("=== Corpus 1: Episodes (04) ===\n\n")
    for word, count in ambiguous1:
        f.write(f"{word}: {count} synsets\n")
    f.write("\n=== Corpus 2: Characters (03) ===\n\n")
    for word, count in ambiguous2:
        f.write(f"{word}: {count} synsets\n")

print("Output written to ambiguity_output.txt")
```
